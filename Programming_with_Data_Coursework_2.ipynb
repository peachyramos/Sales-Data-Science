```python
# Import required modules
import csv
import sqlite3
 
# Connecting to the geeks database
connection = sqlite3.connect('D:\pwdsqlite.db')
 
# Creating a cursor object to execute
# SQL queries on a database table
cursor = connection.cursor()
 
# Table Definition
create_table_marketing = """CREATE TABLE IF NOT EXISTS stores(Name text PRIMARY KEY,
                                        Location text NOT NULL,
                                        MarketingSpend integer);"""
create_table_sales = """CREATE TABLE IF NOT EXISTS sales(Name text PRIMARY KEY,
                                    Year integer NOT NULL,
                                    Area integer NOT NULL,
                                    City text NOT NULL,
                                    Opened integer NOT NULL,
                                    Sales integer NOT NULL)"""        
 
# Creating the table into our
# database
cursor.execute(create_table_marketing)
cursor.execute(create_table_sales)
 
# Opening the person-records.csv file
file_marketing = open('/MarketingSpend2010-2019.csv')
file_sales = open('/YearlySales2010-2019.csv')
 
# Reading the contents of the
# person-records.csv file
contents_marketing = csv.reader(file_marketing)
contents_sales = csv.reader(file_sales)
 
# SQL query to insert data into the
# stores and sales table
insert_records_marketing = "INSERT OR IGNORE INTO stores (Name, Location, MarketingSpend ) VALUES(?, ?, ?)"
insert_records_sales=  "INSERT OR IGNORE INTO sales  (Name, Year, Area, City, Opened, Sales ) VALUES(?, ?, ?, ?, ?, ?)"



# Importing the contents of the file
# into our person table
cursor.executemany(insert_records_marketing, contents_marketing)
cursor.executemany(insert_records_sales, contents_sales)
 
# SQL query to retrieve all data from
# the person table To verify that the
# data of the csv file has been successfully
# inserted into the table
select_all_marketing = "SELECT * FROM stores"
rows1 = cursor.execute(select_all_marketing).fetchall()

select_all_marketing = "SELECT * FROM sales"
rows2= cursor.execute(select_all_marketing).fetchall()

# Output to the console screen
for r in rows1:
    print(r)
for x in rows2:
    print(x)

# Committing the changes
connection.commit()
 
# closing the database connection
connection.close()
```
